# Predicting House Prices

##Fire Up GraphLab Create

```python
import graphlab
```

#Load Some House Sales Data

Dataset is from house sales in King County, the region where the city of Seattle, WA is located

```python
sales = graphlab.SFrame('home_data.gl/')
```

```python
sales
```

#Exploring the Data for Housing Sales

The house price is correlated with the number of square feet of living space

```python
graphlab.canvas.set_target('ipynb')
sales.show(view = "Scatter Plot", x = 'sqft_living', y = 'price')
```

#Create a Simple Regression Model of sqft_living to price

We now split data into training and testing data with 80% observations in training data. This is done randomly, but we see to it that the randomness is deterministic by setting seed = k, where k can be any constant associated with a unique randomness

```python
train_data, test_data = sales.random_split(0.8, seed = 0)
```

```python
# an experiment of mine
train_and_test = sales.random_split(0.8, seed = 0)
type(train_and_test)
```

#Build the Regression Model

```python
sqft_model = graphlab.linear_regression.create(train_data, target = 'price', features = ['sqft_living'])
```

##Evaluate the Simple Model

```python
test_data
```

```python
print test_data['price'].mean()
```

```python
print train_data['price'].mean()
```

```python
print sqft_model.evaluate(test_data)
```

```python
print sqft_model.evaluate(train_data)
```
